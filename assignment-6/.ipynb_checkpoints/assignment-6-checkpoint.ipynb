### Data visualisations

##### Histplot

```python
sns.histplot(x = 'SepalLengthCm', data = df, bins = 20, kde = True)
```

Inference:
- The highest count of sepal lenth is that of 5cm and 6.6 cm.
- Distribution is very midly right skewed.
- Mean values lie around 6 cm.

```python
sns.histplot(x = 'SepalWidthCm', data = df, bins = 20, kde = True)
```

Inference:
- Maximum sepal width count is over 25 and that of 3cm.
- It follows a relatively normal distribution.

```python
sns.histplot(x = 'PetalLengthCm', data = df, bins = 20, kde = True)
```

Inference:
- Distribution is bimodal with two distinct peaks.
- Shows separation between flower species.
- Petal length usually lies between 1.5 to 7 cm.

```python
sns.histplot(x = 'PetalWidthCm', data = df, bins = 20, kde = True)
```

Inference:
- The distribution is also bimodal.
- Most values lie between 0.1 cm and 2.5 cm.
- This feature is highly useful for classification.

##### Boxplot

```python
sns.boxplot(data=df.drop(['Species'], axis = 1))
```

```python
sns.boxplot(data = df, x = 'Species', y = 'SepalLengthCm')
```

```python
sns.boxplot(data = df, x = 'Species', y = 'SepalWidthCm')
```

```python
sns.boxplot(data = df, x = 'Species', y = 'PetalLengthCm')
```

```python
sns.boxplot(data = df, x = 'Species', y = 'PetalWidthCm')
```

### Inference:
- Sepal width shows outliers.
- Sepal length has moderate spread with no major outliers.
- Petal length and width are well-distributed with very few outliers.
- Petal features are more useful for classification.
