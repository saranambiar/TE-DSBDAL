### Load text files

```python
with open("file1.txt", "r", encoding = "utf-8") as f:
    doc1 = f.read()
with open("file2.txt", "r", encoding = "utf-8") as f:
    doc2 = f.read()
```

```python
print(doc1)
```

```python
print(doc2)
```

### Tokenization
This involves the process of breaking down sentences into tokens -> each word separated by a whitespace is considered as a token.

```python
tokens1 = word_tokenize(doc1)
tokens2 = word_tokenize(doc2)
print(tokens1)
print(tokens2)
```

### POS Tagging
This asigns grammatical tags to each word such as:
- Noun: NN
- Verb: VB
- Adjective: JJ

```python
pos1 = pos_tag(tokens1)
pos2 = pos_tag(tokens2)
print(pos1)
print(pos2)
```

### Stopword removal
This removes commonly used stopwords like is, and, the.

```python
stop_words = set(stopwords.words('english'))
filtered1 = [word for word in tokens1 if word.lower() not in stop_words]
filtered2 = [word for word in tokens2 if word.lower() not in stop_words]

print(filtered1)
print(filtered2)
```

### Stemming
This step is used to strip every word down to its root word.
Ex: eating -> eat

```python
stemmer = PorterStemmer()
```

```python
stemmed1 = [stemmer.stem(word) for word in filtered1]
stemmed2 = [stemmer.stem(word) for word in filtered2]
print(stemmed1)
print(stemmed2)
```

### Lemmatization
It involves converting every stemmed word to its closest mappable dictionary word.

```python
lemmatizer = WordNetLemmatizer()
lemma1 = [lemmatizer.lemmatize(word) for word in filtered1]
lemma2 = [lemmatizer.lemmatize(word) for word in filtered2]

print(lemma1)
print(lemma2)
```

### TF-IDF Vectorizer
- Term Frequency: Refers to the number of a times a word appears in a document
- Inverse Document Frequency: How unique/important a word is across documents
- **TF-IDF = TF × IDF**
- It is used to find important words across documents.

```python
documents = [doc1, doc2]
vectorizer = TfidfVectorizer(stop_words = 'english')
tfidf_matrix = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
```

```python
df = pd.DataFrame(tfidf_matrix.toarray(), columns = feature_names)
df
```

```python
for i in range(len(df)):
    print(df.iloc[i].sort_values(ascending=False).head(10))
```
