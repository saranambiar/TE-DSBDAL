## NBA dataset

### Loading dataset

```python
df1 = pd.read_csv("nba.csv")
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1.info()
```

### Data cleaning

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(subset = ['Name', 'Team', 'Number', 'Position', 'Age', 'Height', 'Weight'])
```

```python
df1.isnull().sum()
```

```python
df1['College'] = df1['College'].fillna('Unknown')
```

```python
df1['Salary'] = df1['Salary'].fillna(df1['Salary'].mean())
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df1.duplicated().sum()
```

Thus we handled missing values by dropping wherever there was only 1 missing value. We listed all NaN in College as unknown as trying to impute this categorical column with mode can skew ground truth. Salary was imputed with mean values.
No duplicates found.

### NBA statistics

```python
df1.describe()
```

##### Group by category:

```python
df1_height = df1.groupby('Height')
df1_height.head()
```

```python
df1_height["Weight"].describe()
```

We group data by height and display statistics of weight.

```python
df1_team = df1.groupby('Team')
df1_team.head()
```

```python
df1_team['Salary'].describe()
```

We group data by team and display statistics of salary.

```python
# creating category of age
bins = [19, 25, 31, 36, 40]
labels = ['Young', 'Prime', 'Seasoned', 'Veteran']
df1['AgeGroup'] = pd.cut(df1['Age'], bins = bins, labels = labels)
df1.head()
```

```python
df1_age = df1.groupby('AgeGroup')
df1_age.head()
```

```python
df1_age['Salary'].describe()
```

We group data by age category and display statistics of salary.
