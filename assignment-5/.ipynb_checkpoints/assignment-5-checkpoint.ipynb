# Data Visualisation II - Assignment 5

### Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading dataset


```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Data Cleaning

```python
df.isnull().sum()
```

```python
df.drop(columns = ['Cabin'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df = df.dropna(subset = 'Embarked')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

### Data visualisation

```python
sns.boxplot(x = 'Sex', y = 'Age', hue = 'Survived', data = df)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Distribution of Age wrt Gender')
plt.show()
```

### Inferences:
- The age median of men for survival was almost the same -> around 30 years of age. More number of younger men survived than older men.
- Most women who survived had a median age of about 30 years. The ones who did not survive largely fell under the 20-30 years category.
