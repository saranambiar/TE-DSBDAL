### Data Preprocessing

Checking and handling missing values:

```python
df.isnull().sum()
```

BuildingArea, YearBuilt, Car and CouncilArea have missing values. We handle them by:
- Imputing BuildingArea and YearBuilt with means: too many meaningful samples to drop
- Imputing Car with means
- Imputing CouncilArea with mode: it is a categorical variable, we cannot perform numeric mean or median operations to handle it. Too many meaningful samples to drop.

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
```

```python
df['YearBuilt'] = df['YearBuilt'].fillna(df['YearBuilt'].mean())
```

```python
df['Car'] = df['Car'].fillna(df['Car'].mean())
```

```python
df['CouncilArea'] = df['CouncilArea'].fillna(df['CouncilArea'].mode()[0])
```

mode() returns a series of values (like a list) of most frequently occuring values. [0] ensures that we select the highest occuring value from the same.

### Data Formatting and Normalization

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
```

```python
df['Postcode'] = df['Postcode'].astype(int)
df['Bedroom2'] = df['Bedroom2'].astype(int)
df['Bathroom'] = df['Bathroom'].astype(int)
df['Car'] = df['Car'].astype(int)
df['YearBuilt'] = df['YearBuilt'].astype(int)
df['Propertycount'] = df['Propertycount'].astype(int)
```

```python
df.dtypes
```

We applied type conversion to convert relevant variables to its respective data types.

### Convert categorical variables to numerical

There are 3 ways to convert categorical to numerical values:

##### One Hot Encoding
- We use one hot encoding when there are a smaller number of possible values for categorical variables. It converts each value for a variable to its own variable and assigns a value of '1' if true and '0' if false.

```python
type_encoded = pd.get_dummies(df['Type'], prefix = 'Type')
Regionname_encoded = pd.get_dummies(df['Regionname'], prefix = 'Regionname')
```

##### Dummy Coding
- It is the same as one hot encoding except it drops 1 column - the values of the other columns implicitly signify the value the missing column would have.

```python
Method_encoded = pd.get_dummies(df['Method'], prefix = 'Method', drop_first = True)
```

```python
df = pd.concat([df, type_encoded, Regionname_encoded, Method_encoded], axis = 1)
```

```python
df = df.drop(['Type', 'Regionname', 'Method'], axis = 1)
```

##### LabelEncoder
- This is a preprocessing method from sklearn that converts all values for a categorical variable to its numerical equivalent. This is done to ensure machine learning models can understand the signal from each dataset.

```python
le = LabelEncoder()
for col in ['Suburb', 'Address', 'SellerG', 'CouncilArea', 'Postcode', 'Date']:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```
